```python
# Define a function that computes the mean of the values in a list and prints the results.

def compute_mean(values):
    mean_value = sum(values) / len(values)
    print(mean_value)

# Compute the mean and call output.
compute_mean([2, 4, 6, 8])
    
```

The code above computes the mean of the values in a list and prints the results. The equation of the mean of a list of values is computed as follows:

$$
\bar{X}=\frac{X_{1}+X_{2}+...+X_{n}}{n}=\frac{1}{n}\left(\sum_{i=1}^{n}X_{i}\right)
$$
